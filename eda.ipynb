```python
import json
import pandas as pd

data = json.load(open('data.json', encoding = 'cp850'))
```

```python
no_dup = set()
region_keys = {}
nw = []
ne = []
sw = []
se = []


for objects in data['RECDATA']:
    del objects['CampsiteID']
    del objects['CampsiteName']
    del objects['CreatedDate']
    if objects['FacilityID'] not in no_dup:
        no_dup.add(objects['FacilityID']);
    else:
        objects['FacilityID'] = 'NA'
    if (objects['CampsiteType'] == 'RV NONELECTRIC')|\
    (objects['CampsiteType'] == 'RV ELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP RV AREA NONELECTRIC')|\
    (objects['CampsiteType'] == 'PARKING'):
            objects['CampsiteType'] = "RV";
    if (objects['CampsiteType'] == 'SHELTER NONELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP SHELTER ELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP SHELTER NONELECTRIC')|\
    (objects['CampsiteType'] == 'OVERNIGHT SHELTER NONELECTRIC')|\
    (objects['CampsiteType'] == 'OVERNIGHT SHELTER ELECTRIC')|\
    (objects['CampsiteType'] == 'SHELTER ELECTRIC'):
            objects['CampsiteType'] = "Shelter";
    if (objects['CampsiteType'] == 'GROUP TENT ONLY AREA NONELECTRIC')|\
    (objects['CampsiteType'] == 'TENT ONLY NONELECTRIC')|\
    (objects['CampsiteType'] == 'YURT')|\
    (objects['CampsiteType'] == 'TENT ONLY ELECTRIC')|\
    (objects['CampsiteType'] == 'OVERNIGHT SHELTER ELECTRIC')|\
    (objects['CampsiteType'] == 'BALL FIELD'):
            objects['CampsiteType'] = "Tent";
    if (objects['CampsiteType'] == 'CABIN NONELECTRIC')|\
    (objects['CampsiteType'] == 'CABIN ELECTRIC'):
            objects['CampsiteType'] = "Cabin";
    if (objects['CampsiteType'] == 'BOAT IN')|\
    (objects['CampsiteType'] == 'ANCHORAGE'):
            objects['CampsiteType'] = "Water";
    if (objects['CampsiteType'] == 'LOOKOUT'):
            objects['CampsiteType'] = "View";
    if (objects['CampsiteType'] == 'WALK TO')|\
    (objects['CampsiteType'] == 'GROUP WALK TO')|\
    (objects['CampsiteType'] == 'HIKE TO'):
            objects['CampsiteType'] = "Hike";
    if (objects['CampsiteType'] == 'PICNIC')|\
    (objects['CampsiteType'] == 'GROUP WALK TO'):
            objects['CampsiteType'] = "Picnic";
    if (objects['CampsiteType'] == 'EQUESTRIAN NONELECTRIC')|\
    (objects['CampsiteType'] == 'EQUESTRIAN ELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP EQUESTRIAN'):
        objects['CampsiteType'] = "Horseback";
    if (objects['CampsiteType'] == 'STANDARD NONELECTRIC') |\
    (objects['CampsiteType'] == 'STANDARD ELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP STANDARD ELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP STANDARD NONELECTRIC')|\
    (objects['CampsiteType'] == 'GROUP STANDARD AREA NONELECTRIC')|\
    (objects['CampsiteType'] == 'YES')|\
    (objects['CampsiteType'] == 'GROUP STANDARD AREA ELECTRIC')|\
    (objects['CampsiteType'] == 'MOORING')|\
    (objects['CampsiteType'] == 'MANAGEMENT'):
            objects['FacilityID'] = 'NA';
    if objects['CampsiteLatitude'] > 39.50 and objects['CampsiteLongitude'] < -98.35:
        objects['Region'] = 'north_west';
    elif objects['CampsiteLatitude'] > 39.50 and objects['CampsiteLongitude'] > -98.35:
        objects['Region'] = 'north_east';
    elif objects['CampsiteLatitude'] < 39.50 and objects['CampsiteLongitude'] < -98.35:
        objects['Region'] = 'south_west';
    elif objects['CampsiteLatitude'] < 39.50 and objects['CampsiteLongitude'] > -98.35:
        objects['Region'] = 'south_east';
    if objects['FacilityID'] == 'NA':
        objects['CampsiteType'] = 'NA'
        objects['TypeOfUse'] = 'NA'
        objects['Loop'] = 'NA'
        objects['CampsiteAccessible'] = 'NA'
        objects['CampsiteLongitude'] = 'NA'
        objects['CampsiteLatitude'] = 'NA'
        objects['LastUpdatedDate'] = 'NA'
        objects['Region'] = 'NA';
    objects['Location'] = [objects['CampsiteLatitude'], objects['CampsiteLongitude']]
    del objects['CampsiteLatitude']
    del objects['CampsiteLongitude']
#     fac_id = objects['FacilityID']
#     del objects['FacilityID']
#     final_dict[fac_id] = objects
    if objects['Region'] == 'north_west':
        nw.append(objects)
    if objects['Region'] == 'north_east':
        ne.append(objects)
    if objects['Region'] == 'south_west':
        sw.append(objects)
    if objects['Region'] == 'south_east':
        se.append(objects)
    
```
